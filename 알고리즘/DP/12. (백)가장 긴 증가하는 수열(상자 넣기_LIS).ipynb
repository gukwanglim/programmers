스터디에서 하고 있는 문제 중에서 상자 넣기와 비슷한 문제이다.

각 수까지 갈 때의 최대의 길이를 저장하고, 적절할 때 이를 꺼내어 쓰는 방식으로 문제를 풀 것이다.
맨 처음 숫자는 입력과 관계없이 무조건 0 이 존재한다고 가정하고 '0 2 5 4 6 1 7 8' 이라는 입력이 들어왔다고 하자. (실제 문제라면 0을 제외한 '2 5 4 6 1 7 8'이 되는 것이다.)

각 수가 마지막일 때의 수열 길이 최댓값을 저장해두는 배열을 dp, 비교해줄 수들을 저장하는 배열을 cp라고 한다.
최초값으로 0이 들어가있고 첫 번째 수는 무조건 수열의 길이가 1이다. 
따라서 
    dp = [0, 1]
    cp = [0, 2]    꼴이 된다.   (0이 들어간 상태에서 다음 숫자인 2가 들어온 상황)

이제 5를 비교해보자.
cp에 2가 최대이다. 5는 이보다 크므로 그대로 cp에 저장해주고, dp에 저장된 2의 인덱스는 1이다. 이에 1 만큼 더해주자.
    dp = [0, 1, 2]
    cp = [0, 2, 5]

4를 비교한다. 4는 5보다 작고 2보다 크다. 2의 인덱스는 1 이므로 여기에 1만큼 더해서 저장해주자.
단, 여기서 5를 4로 교체한다. 다음으로 들어올 수들의 수열 길이가 2보다 크려면, 5가 아닌 4와 비교해주면 되기 때문이다.
    dp = [0, 1, 2, 2]
    cp = [0, 2, 4]

연습 코드를 설명하기에 앞서 백준의 문제 상자 넣기를 기준으로 설명하겠다.
문제는 표현하는 단어만 다를 뿐 같은 방식으로 풀 수 있다.

```python
# 연습 코드

nums = [0,2,5,4,6,1,7,8]                    # 실제 문제라면 [2,5,4,6,1,7,8]와 같지만, 인덱스를 맞추기 위해 0을 포함한 리스트의 형태로 받아온다.
n = len(nums)-1

dp = [0, 1]                                 # dp는 인덱스(몇 번째 상자)를 기준으로 몇 개의 상자를 사용한 것인지를 나타낸다.
cp = [0, nums[1]]                           # cp는 현재까지 사용된 상자의 번호를 나타낸다.      먼저, 제일 처음에 위치한 크기 2의 상자를 cp에 넣는다. 이때, 상자 1개를 사용했으므로 dp에 1이 추가된 것이다.

for i in range(2, n+1):                     # 첫 번째 상자를 직접 넣어준 상태에서 두 번째 상자부터는 코드를 사용하여 넣는다.
    for j in range(len(cp)-1,-1,-1):        # 간단하게 설명하자면 nums[i]는 이번에 사용해야할 상자의 크기를 말하고, j는 몇 개의 상자가 들어가는지를 나타낸다.(즉, j+1이 사용한 상자의 개수이다.)
        if nums[i] > cp[j]:                 # 만약, 이번에 사용해야할 상자의 크기가 사용된 상자(역순으로 사용된)보다 클 경우(상자를 사용할 수 있는 경우) dp에 상자를 사용했다는 표시를 넣는다. TODO 1번 내용 추가 확인
            dp.append(j+1)                  # dp에 들어가는 숫자는 현재 인덱스(몇 번째 위치한 상자)에 대해 몇 개의 상자를 사용했는지(j+1)를 나타낸다.

            if j+1 < len(cp):               # 현재 사용한 상자의 개수(j+1)가 사용한 상자의 개수(cp)보다 작을 경우에 현 위치의 cp 값을 변경해준다.   TODO 1번 추가 내용에서 설명된 부분이다.
                cp[j+1]=nums[i]
            else:                           # 그렇지 않다면 이번에 사용하게 될 상자는 아무런 문제없이 사용될 수 있다는 뜻이므로 cp에 사용해야할 상자를 넣어준다.
                cp.append(nums[i])
            print(cp,dp)
            break

# 1. if nums[i] > cp[j]:을 사용할 때, for 구문으로 돌리는 이유는 이번에 사용하게 될 상자의 크기가 이전까지 사용한 상자의 크기보다 작을 경우를 위해 사용한다.
    # 코드의 문제를 예로 들자면 dp = [0, 2, 4, 6], cp = [0, 1, 2, 2, 3]으로 크기가 6인 상자를 사용하는 순간 3개의 상자를 사용한 상황이다.
    # 다음으로 들어올 상자의 크기는 1이다. 즉, 지금까지 사용했던 모든 상자의 크기보다 작은 상자를 사용하는 것이다.
    # 크기 1의 상자를 사용하지 않고 다음 상자를 사용한다하여 최대 사용값이 나올 수 있지만, 지금까지 사용한 상자를 버리고 크기 1의 상자부터 사용했을 때, 훨씬 많은 상자를 사용할 가능성이 존재한다.
    # 때문에 작은 상자의 값으로 갱신이 필요하다.
        # 현재 사용한 상자 중에 제일 작은 상자는 2이다. 또한, 이번에 사용해야할 상자의 크기는 1이다.
        # 즉, 크기가 1인 상자는 하나의 상자부터 숫자를 세야하는 상황이다.
        # 1개의 상자를 사용한 순간의 위치는 현재 사용한 상자 중에서 크기 2인 상자를 사용한 순간이다.
        # 그렇기 때문에 크기 2인 상자를 버리고 1인 상자로 변경하는 것이다.(어차피 크기 2부터 6까지의 상자를 사용한 값은 6에 저장되어 있다고 보면 된다.)
        
        # 현재 예시가 1이어서 크기 2인 상자를 제거하고 1을 넣은 것이다.
        # 만약, 예시가 3이었다면 크기 4인 상자를 제거하고 3이 들어가게 된다.
        # 이 경우, dp는 [0, 1, 2, 2, 3, 2]가 되면서 상자를 바꿔치게 된다. 
```

최종적인 코드는 가장 긴 증가하는 수열 문제와 상자 넣기 문제가 동일하다.

```python
n = int(input())
nums = [0] + list(map(int,input().split()))

dp = [0, 1]
cp = [0, nums[1]]

for i in range(2, n+1):
    for j in range(len(cp)-1,-1,-1):

        if nums[i] > cp[j]:
            dp.append(j+1)

            if j+1 < len(cp):                   # 현재 사용해야할 크기의 상자가 존재할 위치보다 사용된 상자의 수가 많은 경우에 해당 위치의 값을 변경한다.
                cp[j+1]=nums[i]
            else:
                cp.append(nums[i])
            break

print(max(dp))
```
