```python
# 유니온 파인드

def solution(commands):
    answer = []
    
    parent = list([[i, j] for j in range(51)] for i in range(51))
    graph = list(['EMPTY' for _ in range(51)] for _ in range(51))
    
    # find 설명
        # 예를 들어 parent[0][0] = [0, 0]이고 parent[1][1] = [0, 0]일 경우

        # find(0, 0)은 parent[0][0] == [0, 0]이 성립하기 때문에 자기 자신이 부모라는 뜻이 된다. 
        # 즉, find(0, 0)은 parent[0][0]을 return 한다.

        # find(1, 1)은 parent[1][1] == [1, 1]을 성립하지 않기 때문에 if 구문을 타지 않고 parent[1][1]을 정의하게 된다.
        # 즉, parent[1][1] = find(0, 0) (여기서 find()에 들어가는 것은 현재 parent[1][1]이 가지고 있는 값 [0, 0]) 이 되며
        # parent[1][1]의 값은 [0, 0]을 가지고 있는 parent[0][0]이 된다.
        # 쉽게 말해, parent[1][1]의 부모인 parent[0][0]을 찾는 것이다.
    def find(y, x):
        if parent[y][x] == [y, x]:
            return parent[y][x]
        parent[y][x] = find(parent[y][x][0], parent[y][x][1])
        return parent[y][x]
    
    # union 설명
        # 먼저, 들어오게 되는 a, b, c, d는 각각을 (r1, c1), (r2, c2)라고 한다.
        # (r1, c1), (r2, c2)에 대한 find를 실행하여 부모를 찾아낸다.

        # 두 셀이 같은 경우(두 셀의 부모가 같은 경우(여기서 부모는 병합을 의미한다고 볼 수 있다.)) 무시한다는 조건으로 인해 첫 번째 if가 된다.
        # 두 셀 중 한 셀이 값을 가지고 있을 경우 병합된 셀의 값을 갖기 때문에 graph[r1][c1]이 EMPTY인 경우 [c, d]의 값을 가진다.
            # graph[r2][c2]의 값이 EMPTY인 경우를 생략한 것은 다음 조건 때문에
        # 두 셀이 모두 값을 가지고 있다면 graph[r1][c1]의 값을 가지게 된다.

        # 여기서 헷갈리지 말하야 할 것은 graph[r1][c1]의 값을 가진다고 하여 graph[r2][c2]의 값을 변경하는 것이 아닌 위치(parent)만을 활용하는 것
    def union(a, b, c, d):
        a, b = find(a, b)
        c, d = find(c, d)

        if a == c and b == d:
            return
        if graph[a][b] == 'EMPTY':
            parent[a][b] = [c, d]
        else:
            parent[c][d] = [a, b]   
    
    for command in commands:
        li_command = command.split()
        
        if li_command[0] == 'UPDATE' and len(li_command) == 4:
            com, r, c, value = li_command
            y, x = find(int(r), int(c))
            graph[y][x] = value
            
        elif li_command[0] == 'UPDATE':
            com, value1, value2 = li_command
            for i in range(51):
                for j in range(51):
                    y, x = find(i, j)
                    if graph[y][x] == value1:
                        graph[y][x] = value2
                        
        elif li_command[0] == 'MERGE':
            com, r1, c1, r2, c2 = li_command
            union(int(r1), int(c1), int(r2), int(c2))
            
        elif li_command[0] == 'UNMERGE':
            com, r, c = li_command
            y, x = find(int(r), int(c))
            
            save_graph = graph[y][x]
            check = []
            for i in range(51):
                for j in range(51):
                    if [y, x] == find(i, j):
                        check.append([i, j])
                        
            for [y, x] in check:
                parent[y][x] = [y, x]
                graph[y][x] = 'EMPTY'
                
            parent[int(r)][int(c)] = [int(r), int(c)]
            graph[int(r)][int(c)] = save_graph
            
        elif li_command[0] == 'PRINT':
            com, r, c = li_command
            y, x = find(int(r), int(c))
            answer.append(graph[y][x])
        
    return answer
```
