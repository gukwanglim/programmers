부분행렬 합의 최대값을 구하는 문제이다.
부분행렬의 합을 구하기 위해서는 부분합이라는 알고리즘을 사용해야 한다.

![그래프](./image/1-2.png)

![그래프](./image/1-3.png)

사진과 같이 부분합을 구하기 위해서는 하나의 표가 필요하다.

예를 들자면
    2 3 -21 -22 -23     의 값이 존재할 경우 왼쪽부터 차례로 더한 값을 새롭게 작성한다.(작성할 때는 0부터 시작한다.)
    0 2 5 -16 -38 -61     처음 리스트의 인덱스 1번부터 3번까지의 합을 구하고 싶다면 새롭게 작성한 합 리스트의 4번째 인덱스 값과 1번째 인덱스 값을 뻬면 된다.

새롭게 작성한 인덱스의 4번째 값 : -38, 1번째 값 : 5
    -> -38 - 2 = -40

이제 기존 리스트에서 구하고 싶던 값을 계산해보면
인덱스의 3번째 값 : -22, 2번째 값 : -21, 1번째 값 : 3
    -> -22 + -21 + 3 =  -40
